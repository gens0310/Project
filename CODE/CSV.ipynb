```python
PATH = '/content/drive/MyDrive/Data/데이터셋/kfood2/구이'
folder_list = os.listdir(PATH)
print(folder_list)
```

```python
folder_list = ['갈비구이', '고등어구이', '닭갈비', '불고기', '삼겹살']
for folder_name in folder_list:
    data_dir = PATH + '/' + folder_name
    data_list = os.listdir(data_dir)
    for image_file in data_list:
        fname, ext = os.path.splitext(image_file)
        if ext == '.jpg':
            image_name = image_file
            new_data = {'대분류' : '구이', '소분류' : folder_name, '이미지' : image_name}
            df = df.append(new_data, ignore_index = True)
```
