```python
# 메모리 변수 모두 제거
all = [var for var in globals() if var[0] != '_']
for var in all:
    del globals()[var]
```

```python
df = pd.read_csv('통합.csv')
df.info()
```

```python
df = df.drop(['Unnamed: 0'], axis = 1)
df = df.sort_values(by = ['대분류', '소분류', '이미지'])
df = df.reset_index(drop = True)
df = df.astype({'Category' : 'int8'})
```

```python
df = df[18244:22805]
df.tail()
```

# 4. NPZ

```python
X = []
y = []
Large_Category = list(df['대분류'])
Small_Category = list(df['소분류'])
Image_Name = list(df['이미지'])
Label = list(df['Category'])
PATH = '/content/drive/MyDrive/Data/데이터셋/kfood2'
for i in range(len(df)):
    img = Image.open(PATH + '/' + \
                     str(Large_Category[i]) + '/' + \
                     str(Small_Category[i]) + '/' + \
                     Image_Name[i])
    img = img.convert('RGB')
    img = img.resize((64, 64))
    imgarr = np.array(img) / 255
    X.append(imgarr)
    y.append(Label[i])
```

```python
print(np.shape(X))
print(np.shape(y))
print(np.unique(y))
```

```python
type(X), type(y)
```

```python
X = np.array(X, dtype = 'float32')
y = np.array(y, dtype = 'int8')
```
